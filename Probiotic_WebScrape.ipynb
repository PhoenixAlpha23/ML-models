```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# List of company URLs
company_urls = [
    "http://www.nestle.com",
    "http://colacompany.com"
    "http://www.drreddys.com",
    "http://www.pfizer.com",
    "http://www.jnj.com",
    "http://www.danone.com",
    "http://www.bayer.com",
    "http://www.generalmills.com",
    "http://www.kelloggs.com",
    "http://www.unilever.com",
    "http://www.mondelezinternational.com",
    "http://www.theheinekencompany.com",
    "http://www.cardinalhealth.com"

]

# Keywords and attributes
keywords = {
    "Relevant Category": ["food", "beverage", "bulk", "formulations"],
    "Manufacturer": ["manufacturer", "made by", "produced by"],
    "Brand": ["brand", "label"],
    "Distributor": ["distributor", "supplied by"],
    "F&B": ["food", "beverage"],
    "Probiotics": ["probiotic", "probiotics"],
    "Fortification": ["fortify", "fortification"],
    "Gut Health": ["gut health", "digestive health"],
    "Womens Health": ["women's health", "female health"],
    "Cognitive Health": ["cognitive health", "brain health"]
}

# Function to extract relevant keyword-based data from a website
def extract_keyword_data(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        text = soup.get_text().lower()

        # Check for keywords in the text
        result = {"URL": url}
        for attribute, words in keywords.items():
            result[attribute] = "Yes" if any(word in text for word in words) else "No"

        return result
    except Exception as e:
        return {"URL": url, "Error": str(e)}

# Scrape each company website
data = [extract_keyword_data(url) for url in company_urls]

# Create a DataFrame from the extracted data
df = pd.DataFrame(data)

# Save the data to a CSV file
df.to_csv("keyword_analysis.csv", index=False)

print("Scraping complete. Data saved to 'keyword_analysis.csv'.")
```
